```python
import bs4
import requests
```

```python
url = "https://search.shopping.naver.com/search/all?query=%EB%A7%A5%EB%B6%81&frm=NVSHATC"
result = requests.get(url).text
result
```

```python
bsObj = bs4.BeautifulSoup(result, 'html.parser')
```

```python
#'a'는 태그 값을 입력하였고, 키:값 형태는 주로 속성의 id 혹은 class 값으로 검색할때 많이 사용한다.
items = bsObj.find_all('a',{'class':'basicList_link__JLQJf'})
```

```python
for item in items:
  print(item.text)
```
